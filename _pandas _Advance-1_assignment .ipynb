Question - 9 
ans - 

```python
import pandas as pd

df = pd.DataFrame({"Values":[1,2,3,4,5,6,12,13,15,23,57]})

def calculation(df):
    col_val = df['Values']
    
    mean_val = col_val.mean()
    median_val = col_val.median()
    deviation = col_val.std()
    
    return mean_val , median_val , deviation
```

```python
result = calculation(df)
result
```

```python
import pandas as pd
date = ['2023/06/02','2023/06/09','2023/06/16','2023/06/23','2023/06/30','2023/07/06',
       '2023/07/13','2023/07/20','2023/07/27','2023/08/04']
df = pd.DataFrame({"Sales":[100,25,35,50,78,69,80,48,58,60],
                  "Date":date})
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df
```

```python
def moving_average(df):
    df['Moving_Average'] = df['Sales'].rolling(window = 7, min_periods = 1).mean()
    return df
```

```python
obj_moving_average = moving_average(df)
obj_moving_average
```

```python
df = pd.DataFrame({"Date":pd.date_range(start = '2023/01/01' , end = '2023/02/02')})
```

```python
df
```

```python
def select_rows(df):
    
    start_date = '2023/01/01'
    end_date = '2023/01/31'
    
    selected_rows = df[(df['Date'] >= start_date) & (df['Date']<= end_date)]
    
    
    
    
    return selected_rows

obj_select_rows = select_rows(df)
```
